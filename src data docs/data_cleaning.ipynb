```python
import pandas as pd

df = pd.read_csv('../src data docs/iris.csv')

# Check missing values
print(df.isnull().sum())

# Fill numeric columns with median (safe strategy)
df.fillna(df.median(numeric_only=True), inplace=True)

# Verify
print(df.isnull().sum())
```

```python
print("Duplicates before:", df.duplicated().sum())

df.drop_duplicates(inplace=True)

print("Duplicates after:", df.duplicated().sum())
print("New shape:", df.shape)
```

```python
# Check current types
print(df.dtypes)

# Species should be categorical
df['Species'] = df['Species'].astype('category')

# Verify
print(df.dtypes)
print(df['Species'].unique())  # see all 3 species names
```
